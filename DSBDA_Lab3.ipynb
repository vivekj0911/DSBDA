```python
#identifying outliers using boxplot
col = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
virginica.boxplot(column=col)
```

```python
#identifying outliers using interQuantileRange method
sorted_SepalWidth = sorted(virginica['Sepal_Width'])
sorted_SepalWidth
```

```python
q1 = np.percentile(sorted_SepalWidth,25)
q3 = np.percentile(sorted_SepalWidth,75)
print(q1,q3)
```

```python
IQR = q3 - q1
IQR
```

```python
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print(lwr_bound, upr_bound)
```

```python
r_outliers = []
for i in sorted_SepalWidth:
  if (i<lwr_bound or i>upr_bound):
    r_outliers.append(i)
print(r_outliers)
```

```python
virginica.boxplot(['Sepal_Width'])
```

```python
virginica_no_outliers = virginica[~virginica['Sepal_Width'].isin(r_outliers)]

virginica_no_outliers.boxplot(['Sepal_Width'])
```

```python
import matplotlib.pyplot as plt
virginica_no_outliers['Sepal_Width'].plot(kind = 'hist', edgecolor = "black")
```

```python
virginica['log_sepalWidth'] = np.log10(virginica['Sepal_Width'])
virginica['log_sepalWidth'].plot(kind = 'hist', edgecolor= 'black')
```
