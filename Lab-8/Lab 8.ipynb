```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic
```

```python
#distribution of age
import seaborn as sns
sns.distplot(x = titanic['age'], bins = 10)
```

```python
sns.jointplot(x = titanic['age'], y = titanic['fare'], kind =
'scatter')
```

```python
sns.rugplot(titanic['fare'])
```

```python
sns.barplot(x='sex', y='age', data=titanic)
```

```python
sns.countplot(x='sex', data=titanic)
```

```python
import matplotlib.pyplot as plt

for col in titanic.select_dtypes(include=np.number):
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=titanic[col])
    plt.title(f'Box Plot of {col}')
    plt.show()
```

```python
titanic = sns.load_dataset('titanic')
for col in titanic.select_dtypes(include=np.number):
    plt.figure(figsize=(8, 6))
    sns.violinplot(x=titanic[col])
    plt.title(f'Violin Plot of {col}')
    plt.show()
```

```python
sns.stripplot(x='sex', y='age', data=titanic, jitter=False)
```

```python
sns.stripplot(x='sex', y='age', data=titanic, jitter=True)
```

```python
sns.stripplot(x='sex', y='age', data=titanic, jitter=True, hue='survived')
```

```python
sns.swarmplot(x='sex', y='age', data=titanic)
```

```python
sns.swarmplot(x='sex', y='age', data=titanic, hue='survived')
```

```python

numerical_features = titanic.select_dtypes(include=np.number)

corr_matrix = numerical_features.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.show()
```
