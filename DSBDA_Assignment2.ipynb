```python
from scipy import stats

data['Math_Score_zscore'] = np.abs(stats.zscore(data['Math_Score']))

threshold = 1.5

print(data['Math_Score_zscore'])


outliers = data[data['Math_Score_zscore'] > threshold]

print(outliers[['Math_Score', 'Placement_Offer_Count']])
```

```python
sorted_wscore = sorted(data['Writing_Score'])
print(sorted_wscore)
```

```python
q1 = np.percentile(data['Writing_Score'], 25)
q3 = np.percentile(data['Writing_Score'], 75)
iqr = q3 - q1
print(q1,q3)
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print(lower_bound)
print(upper_bound)
```

```python
r_outliers = []
for i in sorted_wscore:
  if i < lower_bound or i > upper_bound:
    r_outliers.append(i)
print(r_outliers)
```
